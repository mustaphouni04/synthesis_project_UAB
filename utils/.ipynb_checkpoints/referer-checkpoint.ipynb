```python
import nltk
import re
import sqlite3
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from string import punctuation
from keras.preprocessing import text
from urllib.parse import unquote
from nltk.tokenize import word_tokenize
from nltk.corpus import PlaintextCorpusReader
from gensim.models import Word2Vec
import multiprocessing
physical_devices = tf.config.list_physical_devices('GPU')
# physical_devices

# "../example.csv"
```

```python
def preprocess_referer(ref):
    ref = re.sub(r'[/.-:_=;,]', ' ', ref)
    ref = re.sub(r'-', ' ', ref)
    tokens = word_tokenize(ref)
    return tokens

# load data
conn = sqlite3.connect('data_ref.db')
df.to_sql('data_ref', conn, index=False, if_exists='replace')

# tokenize referers and create corpus
corpus_path = "corpus"
ref_tokens = []

chunk_size = 10000
for chunk in pd.read_sql_query('SELECT referer FROM data_ref', conn, chunksize=chunk_size):
    refs = chunk['referer']
    for ref in refs:
        decoded_ref = unquote(ref)
        tokens = preprocess_referer(decoded_ref)
        ref_tokens.append(tokens)

conn.close()

# save corpus to file
corpus_raw = '\n'.join([' '.join(tokens) for tokens in ref_tokens])
with open(corpus_path + "/referer_corpus_preprocessed.txt", "w", encoding="utf-8") as f:
    f.write(corpus_raw)

print("Corpus created successfully.")
```

```python
print(ref_tokens[:25])
```

```python
EMB_DIM = 30

w2v = Word2Vec(sentences=ref_tokens, vector_size=EMB_DIM, window=5, min_count=5, negative=15,
               epochs=10, workers=multiprocessing.cpu_count())
```

```python
word_vectors = w2v.wv
result = word_vectors.similar_by_word("smiley")
print("Most similar to 'smiley':\n", result[:3])

result = word_vectors.similar_by_word("erleben")
print("Most similar to 'erleben':\n", result[:3])
```

```python
# get the vocabulary
vocab = list(word_vectors.key_to_index.keys())
print(vocab)
# get the word vectors
word_vecs = [word_vectors[word] for word in vocab]
# convert word_vecs to a numpy array
word_vecs_np = np.array(word_vecs)
#print(word_vecs_np[4])
```

```python
# use t-SNE to reduce dimensionality to 2D
tsne = TSNE(n_components=2, random_state=42)
word_vecs_2d = tsne.fit_transform(word_vecs_np)

# plot
plt.figure(figsize=(10, 8))
plt.scatter(word_vecs_2d[:, 0], word_vecs_2d[:, 1], alpha=0.5)

# annotate points with words
for word, (x, y) in zip(vocab, word_vecs_2d):
    plt.annotate(word, (x, y), size=3)

plt.title('t-SNE Visualization of Word Embeddings')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.grid(True)
plt.show()
```

```python
# Function to generate nested list representation for each observation
def generate_embedding_vectors(referer_tokens, word_vectors, emb_dim):
    embedding_vectors = []
    for token in referer_tokens:
        if token in word_vectors:
            embedding_vectors.append(word_vectors[token])
        else:
            embedding_vectors.append([0.0] * emb_dim)  # If token not found, fill with zeros
    return embedding_vectors

# Create a new list to hold the embedding vectors
embedding_vectors_list = []

# Iterate over each observation in the dataframe
for i, row in df.iterrows():
    referer = row['referer']
    decoded_referer = unquote(referer)
    tokens = preprocess_referer(decoded_referer)
    embedding_vectors = generate_embedding_vectors(tokens, word_vectors, EMB_DIM)
    embedding_vectors_list.append(embedding_vectors)

# Add the list of embedding vectors as a new column to the dataframe
df['embedding_vectors'] = embedding_vectors_list

# Display updated dataframe
print(df.head(50))
```
